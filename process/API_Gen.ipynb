```python
import json
import random
import openai
from openai import OpenAI
from tqdm import tqdm
import pandas as pd
```

```python
api_key = "sk-0Ufch49pPjALERG9Fb8c311cAa964aCbB3244dBe05628981"
base_url = 'https://chatapi.onechat.fun/v1/'
client = OpenAI(api_key=api_key, base_url=base_url)
```

```python
def query_gpt(query, prompt, model="gpt-4-1106-preview", temperature=0.3):
    response = client.chat.completions.create(
        model=model,
        #response_format={ "type": "json_object" },
        messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content": question}
        ],
        temperature=temperature
    )
    return response
```

```python
prompt='''
你是顶尖的心理咨询师，患者将向你倾诉它们的苦恼，你的回答需要具备如下的特征：

专业性：你可以用特定的专有名词概括他们们的问题，并且用专业术语回答他们
沟通技巧：优秀的倾听和表达能力，并适时反问以获取更多细节
同理心：能站在咨询者角度理解他们的困扰，提供关怀和支持
洞察力：你的回答具备洞察力，而不是泛泛而谈说空话。
'''
```

```python
exits_instruction=pd.read_json('/root/autodl-tmp/dataset/OESD-GG-zh_cn-1/api_gen.jsonl', lines=True).User.tolist()
```

```python
start_line = 0  # 从第800行开始
count = 0

with open('/root/autodl-tmp/dataset/OESD-GG-zh_cn-1/single_query.jsonl') as file, open('/root/autodl-tmp/dataset/OESD-GG-zh_cn-1/api_gen.jsonl', 'a') as outfile:
    for line_num, line in enumerate(tqdm(file)):
        if line_num < start_line:  # 跳过已经处理过的行
            continue

        entry=dict()
        question=json.loads(line)['User']
        if question in exits_instruction:
            print('already exist, pass')
            continue
        try:
            print('generating')
            response = query_gpt(question, prompt, model='gpt-4-1106-preview', temperature=random.uniform(0.15, 0.25))
            response = response.choices[0].message.content
            print(question)
            print(response)
            count+=1
            if count >= 400:  # 如果已经生成了800个，就停止
                break
        except:
            continue
        entry['User']=question
        entry['Assistance']=response
        
        json.dump(entry, outfile, ensure_ascii=False)
        outfile.write('\n')
```
